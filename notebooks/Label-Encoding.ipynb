```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../data/new_train.csv')
test = pd.read_csv('../data/new_test.csv')
```

```python
train.head(1)
```

```python
test.head(1)
```

```python
all_data = pd.concat([train, test], sort=False)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
all_data['Species'] = label.fit_transform(all_data['Species'])
all_data['Trap'] = label.fit_transform(all_data['Trap'])
```

```python
to_drop = ['Address',
           'Street',
           'AddressNumberAndStreet',
           'AddressAccuracy',
           'NumMosquitos']
all_data.drop(labels=to_drop, axis=1, inplace=True)
```

```python
all_data['Date'] = pd.to_datetime(all_data['Date'])
all_data.set_index('Date', inplace=True)
```

```python
all_data.head()
```

```python
train = all_data[all_data['Id'].isnull()]
test = all_data[pd.notnull(all_data['Id'])]
```

```python
train.drop('Id', axis=1, inplace=True)
```

```python
test.drop('WnvPresent', axis=1, inplace=True)
```

```python
test.isnull().sum()
```

## Save Label Encoded Data Back to CSV

```python
train.to_csv('../data/new_train.csv')
test.to_csv('../data/new_test.csv')
```
