## Visuals

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

% matplotlib inline
```

```python
train = pd.read_csv('../data/new_train.csv', parse_dates=['Date'], index_col='Date')
```

```python
train.columns
```

```python
mosq_over_time = train.groupby(by=train.index).sum()[['NumMosquitos']]
mosq_over_time.head()
```

```python
plt.figure(figsize=(50, 50))
plt.plot(mosq_over_time['2007'])
plt.title('Number of Mosquitos in 2007')
plt.savefig('../images/Number of Mosquitos in 2007.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10,8))
plt.plot(mosq_over_time['2009'])
plt.title('Number of Mosquitos in 2009')
plt.savefig('../images/Number of Mosquitos in 2009.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10,8))
plt.plot(mosq_over_time['2011'])
plt.title('Number of Mosquitos in 2011')
plt.savefig('../images/Number of Mosquitos in 2011.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10,8))
plt.plot(mosq_over_time['2013'])
plt.title('Number of Mosquitos in 2013')
plt.savefig('../images/Number of Mosquitos in 2013.png', bbox_inches='tight')
```

```python
train['Tavg_s1'] = train['Tavg_s1'].astype('int64')
train['Tavg_s2'] = train['Tavg_s2'].astype('int64')
plt.figure(figsize=(10,8))
plt.plot(train['2007'][['Tavg_s1', 'Tavg_s2']])
```

```python
plt.figure(figsize=(10, 8))
sns.distplot(train['WnvPresent'], kde=False, bins=3)
plt.xticks([.165, .825], ['Negative', 'Positive'])
plt.xlabel('')
plt.title('Presence of West Nile Virus')
plt.savefig('../images/Presence of West Nile Virus.png', bbox_inches='tight')
```
