```python
import pandas as pd
```

```python
preds = pd.read_csv('../submissions/XGBClassifier_unscaled_submission.csv')
```

```python
df = pd.read_csv('../data/new_test.csv')
```

```python
df.Id = df.Id.astype('int32')
```

```python
results = pd.merge(preds, df, on='Id')
```

```python
results['Date'] = pd.to_datetime(results['Date'])
```

```python
results.set_index('Date', inplace=True)
```

```python
results_14 = results['2014']
```

```python
at_risk_14 = results_14[results_14['WnvPresent'] > 0.5]
```

```python
at_risk_14.sort_values('WnvPresent', ascending=False)
```

```python
at_risk_14['Date'] = at_risk_14.index
```

```python
at_risk_14['Probability of WNV'] = at_risk_14['WnvPresent']
```

```python
risk_by_day_14 = at_risk_14[['Date', 'Trap', 'Probability of WNV', 'Latitude', 'Longitude']].groupby(at_risk_14.Trap)
```

```python
risk_by_day_14.mean().to_csv('../data/risk_by_trap_2014.csv', index=False)
```
